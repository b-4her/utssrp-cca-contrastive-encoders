```python
import numpy as np
import matplotlib.pyplot as plt

n = 50
p = 200
rho = 0.8

i, j = np.indices((p, p))
Sigma_true = rho ** np.abs(i - j)

L = np.linalg.cholesky(Sigma_true)
Z = np.random.randn(n, p)
X = Z @ L.T

Sample_Sigma = (X.T @ X)/n

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

im1 = axes[0].imshow(Sigma_true, cmap='coolwarm', vmin=-1, vmax=1)
axes[0].set_title(r"True Population Covariance $\Sigma$")
axes[0].set_xlabel("Index j")
axes[0].set_ylabel("Index i")
fig.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)

im2 = axes[1].imshow(Sample_Sigma, cmap='coolwarm', vmin=-1, vmax=1)
axes[1].set_title(r"Empirical Covariance $\hat{\Sigma}$ ($n=50, p=200$)")
axes[1].set_xlabel("Index j")
axes[1].set_ylabel("Index i")
fig.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```
